```python
import pandas as pd
import json
```

```python
df = pd.read_csv("./Downloads/RegPep23 final program datasheet - Sheet9.csv")
actual_df = df[0:155]
actual_df = actual_df.drop(["Unnamed: 0"], axis=1)
actual_df = actual_df.dropna(how='all')
actual_df['Day'] = actual_df['Day'].astype(str)
actual_df = actual_df.rename(columns={"Event category": "eCategory", "Event ID": "eID", "Event title": "eTitle", "First name": "firstName", "Last Name": "lastName", "Talk title": "talkTitle", "City or State": "state", "Role.1": "professionalRole"})
regpepProgram = open("regpepProgramme.json", "w")
actual_df.reset_index().to_json(regpepProgram, orient='records')
# print(json.dumps(json.loads(actual_df.reset_index().to_json(orient='records')), indent=2))
```
